#### 2. Display all the image links from Walt Disney wikipedia page.
Hint: use `.get()` to access information inside tags. Check out the documentation.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
response = requests.get(url, headers=headers)
response
````

````python
soup = BeautifulSoup(response.content, "html.parser")
images = soup.find_all('img')
img_links = []
for img in images:
    link = img.get('src')  # .get() fetches the value of the 'src' attribute
    img_links.append(link)
print(img_links)
````

#### 2.1. List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
response = requests.get(url, headers=headers)
response
````

````python
response.content
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
type(soup)
print(soup.prettify())  # This formats the HTML in a readable way
````

````python
for lang_box in soup.find_all("a", class_="link-box"):
    lang_name = lang_box.find("strong").get_text(strip=True)
    article_count = lang_box.find("small").get_text(strip=True)
    print(lang_name, "-", article_count)
````
